```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
train = pd.read_csv("../input/train.csv")
submission = pd.read_csv("../input/test.csv")
sample_submission = pd.read_csv("../input/sample_submission.csv")
greeks = pd.read_csv("../input/greeks.csv")
```

```python
merged_train = pd.merge(train, greeks, on='Id')
merged_train.info()
```

```python
## 日付がunknownのデータが存在するため以下の前処理をする
## "Epsilon_is_null"のカラムを新しく作成する
## EpsilonのUnknownを別の日付(極端な日付)上書き
merged_train["Epsilon_is_null"] = merged_train["Epsilon"]=="Unknown"
merged_train.loc[merged_train["Epsilon"]=="Unknown", "Epsilon"] = "1/1/1900"
merged_train["Epsilon"] = pd.to_datetime(merged_train["Epsilon"], format='%m/%d/%Y')
```

```python
merged_train
```

```python
## AlphaがAのデータはクラス0に分類され、AlphaがB,D,Gのデータはクラス１に分類される
## そのためAlphaとその他の説明変数の関係を調べる必要がある
## GammaはAlphaの予測に使えそう
## 日付データは年月日を数字に変換。月、日をそれぞれ三角関数で表現https://info.drobe.co.jp/blog/engineering/ml-feature-engineering-date
```

```python
qualitative_variable = ["EJ", "Beta", "Gamma", "Delta", "Epsilon", "Epsilon_is_null"]
qualitative_train = pd.concat([merged_train[qualitative_variable], merged_train[["Alpha"]]], axis=1)
```

```python
pd.crosstab(qualitative_train["Alpha"], qualitative_train["EJ"])
```

```python
pd.crosstab(qualitative_train["Alpha"], qualitative_train["Beta"])
```

```python
pd.crosstab(qualitative_train["Alpha"], qualitative_train["Gamma"])
```

```python
pd.crosstab(qualitative_train["Alpha"], qualitative_train["Delta"])
```

```python
pd.crosstab(qualitative_train["Alpha"], qualitative_train["Epsilon"])
```

```python
date_non_null_train = qualitative_train[qualitative_train["Epsilon_is_null"]==False]
```

```python
tmp = date_non_null_train.groupby(["Epsilon","Alpha"]).size().reset_index()
```

```python
sns.relplot(x=tmp["Epsilon"], y=tmp[0], 
            kind="line", hue=tmp["Alpha"],
            markers=True, dashes=False)
```
